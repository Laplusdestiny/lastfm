```python
import requests
import pandas as pd
from tqdm.auto import tqdm
import datetime
from logging import basicConfig, INFO, info, error

basicConfig(
    filename="get_data.log",
    level=INFO,
    format="%(asctime)s - %(levelname)s:%(name)s - %(message)s"
)
```

```python
host = "http://ws.audioscrobbler.com/2.0/"
header = {"Content-type": "application/json"}
api_key = input("Input API key")
```

```python
def get_result(params, host=host, header=header):
    result = requests.get(
        host,
        headers=header,
        params=params
    )
    return result
```

```python
params = {
    "api_key": api_key,
    "method": "user.getWeeklyTrackChart",
    "user": "Laplusdestiny",
    "format": "json"
}
```

```python
result = get_result(params)
result.status_code
```

```python
result.json()
```

```python
ts = result.json()["weeklytrackchart"]["@attr"]
print(ts)
start = datetime.datetime.fromtimestamp(int(ts["from"]))
end = datetime.datetime.fromtimestamp(int(ts["to"]))
start, end
```

```python
data = list()
for d in tqdm(result.json()["weeklytrackchart"]["track"]):
    data.append([
        d["name"],
        d["artist"]["#text"],
        d["playcount"],
        d["image"][0]["#text"]
    ])
data = pd.DataFrame(data, columns=["name", "artist", "playcount", "img_url"])
data
```
